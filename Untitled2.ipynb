```python
def translate_live_audio(audio_data, lang, state="", detected_language=None):
    # Initialize the OpenAI client
    if detected_language is None:
        result = lang_model.transcribe(audio_data, verbose=False)
        detected_language_code = result['language']
        # Convert the detected language code to language name using the reverse mapping
        detected_language = code_to_language.get(detected_language_code, "Unknown Language")
    # Open the audio file in binary read mode
    output = lang_model.transcribe(audio_data, task="translate")
    text = output["text"]
    encoded_hi = tokenizer(text, return_tensors="pt").to(device)
    generated_tokens = model.generate(**encoded_hi, forced_bos_token_id=tokenizer.get_lang_id(lang))
    output = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)

    state += output[0] + " "
    return detected_language, state, state
```

```python
import librosa

def load_audio(audio_path):
    audio, sr = librosa.load(audio_path, sr=None)  # Load the audio file
    return audio
```

```python
import gradio as gr

from datetime import datetime

# Assuming 'language_map', 'translate_live_audio', and 'translate_and_synthetize' are defined elsewhere.
chat_history = ""  # Initialize chat history as an empty string
structured_chat_history = []  # Initialize structured chat history as an empty list

def append_to_history(message, speaker):
    global chat_history, structured_chat_history
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    chat_history += f"({timestamp}) {speaker}: {message}\n"
    # Append to structured chat history
    structured_chat_history.append({"timestamp ": timestamp, "speaker": speaker, "message": message})
    return chat_history

def transcribe_and_translate_live(audio_path, target_lang_name, state=""):
    lang_code = language_map_1[target_lang_name]
    detected_language, new_state, translation = translate_live_audio(audio_path, lang_code, state)
    # Add to chat history
    speaker = "Caller" if detected_language == target_lang_name else "Caller"
    new_history = append_to_history(translation, speaker)
    return detected_language, new_state, translation, new_history

def transcribe_and_translate(audio_path, target_lang_name):
    lang_code = language_map_2[target_lang_name]
    detected_language, new_state, translation, audio_message = translate_and_synthetize(audio_path, lang_code)
    # Add to chat history
    speaker = "Caller" if detected_language == target_lang_name else "Operator"
    new_history = append_to_history(translation, speaker)
    return detected_language, translation, audio_message, new_history


with gr.Blocks(theme=gr.themes.Default(primary_hue="red", secondary_hue="pink")) as demo:
    with gr.Row():
        with gr.Column():
            # Components for live translation
            audio_input_live = gr.Audio(sources=["microphone"], type="filepath", label="Insert your user's record")
            language_dropdown_live = gr.Dropdown(choices=list(language_map_1.keys()), label="Select Language")
            detected_language_output_live = gr.Textbox(label="Detected Language")
            translation_output_live = gr.Textbox(label="Real-time Translation", lines=3)
            history_textbox = gr.Textbox(label="Conversation History", value=chat_history, lines=10, interactive=False)
            css="footer{display:none !important}"
            audio_input_live.change(fn=transcribe_and_translate_live, inputs=[audio_input_live, language_dropdown_live], outputs=[detected_language_output_live, translation_output_live, history_textbox])

        with gr.Column():
            # Components for controlled translation with submit button
            audio_input = gr.Audio(sources=["microphone"], type="filepath", label="Insert your operator's record")
            language_dropdown = gr.Dropdown(choices=list(language_map_2.keys()), label="Select Language")
            detected_language_output = gr.Textbox(label="Detected Language")
            translation_output = gr.Textbox(label="Translation", lines=3)
            audio_output = gr.Audio(label="Audio Translation", type="numpy", autoplay=True)
            css="footer{display:none !important}"
            audio_input.change(fn=transcribe_and_translate, inputs=[audio_input, language_dropdown], outputs=[detected_language_output, translation_output, audio_output, history_textbox])


demo.dependencies[0]["show_progress"] = "hidden"
demo.launch(debug=True, share=True)
```
